#### One-Hot Encoding

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

df['type_of_meal_plan'] = label_encoder.fit_transform(df['type_of_meal_plan'])
df['room_type_reserved'] = label_encoder.fit_transform(df['room_type_reserved'])
df['market_segment_type'] = label_encoder.fit_transform(df['market_segment_type'])
df['avg_price_per_room'] = label_encoder.fit_transform(df['avg_price_per_room'])
df['booking_status'] = label_encoder.fit_transform(df['booking_status'])
```

```python
#Declare x & y
```

```python
x = df.drop('booking_status', axis=1)
y = df['booking_status']
```

```python
df.dtypes
```

```python
# Data Spliting into training & testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from imblearn.combine import SMOTETomek
```

```python
smpl = SMOTETomek()
```

```python
x_train_os, y_train_os = smpl.fit_resample(x_train, y_train)
```

```python
y_train_os.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_standard = scaler.fit_transform(x)
```

### Logistic  Regression

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
```

```python
lor_mdl = lor.fit(x_train_os, y_train_os)
lor_mdl
```

```python
lor_y_pred = lor_mdl.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, precision_score,recall_score, accuracy_score
```

```python
accuracy_score(lor_y_pred, y_test)
```

```python
cm = confusion_matrix(lor_y_pred, y_test)
```

```python
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
print(classification_report(lor_y_pred, y_test))
```

```python
fpr,tpr,_ = roc_curve(lor_y_pred, y_test)
```

```python
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
```
